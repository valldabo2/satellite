```python
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt

api = SentinelAPI(user='s5pguest', password='s5pguest', api_url='https://s5phub.copernicus.eu/dhus')
```

```python
footprint = geojson_to_wkt(read_geojson('../glass-fire/map.geojson'))

products = api.query(
    footprint, 
    date=('20200920', "20201001"), 
    platformname='Sentinel-5',
    producttype='L2__CH4___'
)

product_df = api.to_dataframe(products)
product_df.head()
```
